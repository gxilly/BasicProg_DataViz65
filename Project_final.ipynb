````python
#price_min 
print(len(opendata_project[opendata_project['price_min'].isnull()])) # ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô missing ‡πÉ‡∏ô price_min
opendata_project = opendata_project.dropna(subset=['price_min'])
````

````python
#subdistrict_id
print(len(opendata_project[opendata_project['subdistrict_id'].isnull()])) # ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô missing ‡πÉ‡∏ô subdistrict_id
opendata_project = opendata_project.dropna(subset=['subdistrict_id'])
````

````python
opendata_project.isnull().any() #check missing ‡∏´‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ clean data ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏¢‡∏±‡∏á‡∏°‡∏µ missing ‡∏≠‡∏¢‡∏π‡πà‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
````
