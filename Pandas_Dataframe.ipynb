- NaN = Not a number
- count --> นับจำนวนแถว แต่ที่ไม่เท่ากัน แปลว่า จาก shape มี 42625 แถว แต่ที่บางแถวไม่ครบ เพราะมันไม่นับค่าว่าง(ค่าว่าง = NaN)
- mean --> ค่าเฉลี่ย (44 --> จากบ้านสี่หมื่นกว่าหลังมีพื้นที่เฉลี่ย 44 ตร.วา)
- std --> ค่า Standard Deviation (SD) หรือค่าเบี่ยงเบนมาตรฐาน ค่านี้มีความสำคัญโดยจะใช้บอกค่าเฉลี่ยของระยะห่างของข้อมูลแต่ละตัวจากค่าเฉลี่ย โดยจะเปรียบเทียบกับค่า mean หรือค่าเฉลี่ยร่วมด้วย หากค่า std ต่ำ แปลว่า ค่าของข้อมูลแต่ละตัวไม่ต่างกันมาก หากค่า std สูง แปลว่า ค่าของข้อมูลแต่ละตัวมีความต่างกันมาก มีการกระจายของข้อมูลสูง (กระจายไประหว่าง 44 + - 21 เวลาสุ่มเจอบ่อยมาก)
- min --> ต่ำที่สุด,น้อยสุด (ค่าพื้นที่ที่น้อยที่สุด คือ 1 ตารางวา)
- 25% --> 25% ของบ้านทั้งหมด มีพท.น้อยกว่า 23 ตร.วา (25%ของจำนวนบ้านที่สำรวจมาทั้งหมดสี่หมื่นกว่าหลังมีพื้นที่น้อยกว่า 23ตรวา หรือมีจำนวนพื้นที่มากที่สุดคือ 23 ตรว )
- 50% --> 50% ของบ้านทั้งหมด มีพท.น้อยกว่า 50 ตร.วา
- 75% --> 75% ของบ้านทั้งหมด มีพท.น้อยกว่า 60 ตร.วา (ไม่ต่ำกว่า)
- max --> สูงสุด,มากสุด (ค่าพท.บ้านที่มากที่สุด คือ 335 ตร.วา)
- price  e+08 = 10^8แพงสุด 58 ล้าน
- parking มีมากสุด 10 ที่
