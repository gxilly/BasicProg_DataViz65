## 2.พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอนของทรัพย์สินที่แพงที่สุดแต่ละชนิด

```python
dm = unit.dropna(subset=['propertytype_name_en','price_min','area_total_min_wa','area_usable_min','count_room_bed','count_parking','count_room_bath','price_min']) # ลบ missing ออก 
```

```python
dm.groupby(['propertytype_name_en']).max()[['area_total_min_wa','area_usable_min','count_room_bed','price_min']]
 #Groupby ชนิดของบ้านแต่ละชนิด ใช้.max() หาค่าแพงที่สุด ใช้list เลือกเฉพาะ column พื้นที่และจำนวนห้องนอน
```

### ของจาร

```python
unit.groupby('propertytype_name_en') [['price_min','area_total_min_wa','area_usable_min','count_room_bed']].max()
```

```python
maxprice = unit.groupby('propertytype_name_en') [['price_min','area_total_min_wa','area_usable_min','count_room_bed']].max().reset_index()
maxprice
```

```python
for row in maxprice.iterrows():
    print(row[1]['propertytype_name_en'],row[1]['price_min'])
    print(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']])
```

ปริ้นค่าในแต่ละแถวของตาราง

```python
unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']]
```

## Dictionary-RowOriented

```python
sales = [{'account' : 'Jones LLC','Jan' : 150,'Feb' : 200,'Mar' : 140},
            {'account' : 'Alpha Co','Jan' : 200,'Feb' : 210,'Mar' : 215},
            {'account' : 'Blue Inc','Jan' : 50,'Feb' : 90,'Mar' : 95}]
```

```python
df = pd.DataFrame(sales)
df
```

## Dictionary-ColumnOriented

```python
sales = {'account' : ['Jones LLC', 'Alpha Co', 'Blue Inc'],
            'Jan' : [150,200,140],
            'Feb' : [200,210,90],
            'Mar' : [140,215,95]}
```

```python
df = pd.DataFrame.from_dict(sales)
df
```

## List-RowOriented

```python
sales = [['Jones LLC', 150, 200, 140],
            ['Alpha Co', 200, 210, 215],
            ['Blue Inc', 50, 90, 95]]
columnname = ['account','Jan','Feb','Mar']
```

```python
df = pd.DataFrame.from_records(sales,columns=columnname)
df
```
